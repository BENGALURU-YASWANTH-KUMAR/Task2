# Collabarative Filtering Based On Recommender System

# üìö Advanced Book Recommendation System: Unleashing the Power of Collaborative Filtering

## üîç Data Processing and Filtering

### 1. Identifying Active Readers üë•
```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
daily_readers = x[x].index
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(daily_readers)]
```
This step focuses on active users who have rated more than 200 books, ensuring high-quality data for our model.

### 2. Selecting Popular Books üìä
```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```
We further refine our dataset by including only books with at least 50 ratings, reducing noise and focusing on well-established titles.

### 3. Creating a User-Item Matrix üî¢
```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt.fillna(0, inplace=True)
```
This pivotal step transforms our data into a matrix where rows represent books, columns represent users, and cells contain ratings.

## üß† Collaborative Filtering Algorithm

### 1. Calculating Similarity Scores üî¨
```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(pt)
```
We use cosine similarity to measure how similar books are based on user ratings. This is the core of our collaborative filtering approach.

### 2. Recommendation Function üéØ
```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:7]
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```
This function finds the most similar books to a given title and returns their details, including title, author, and image URL.
